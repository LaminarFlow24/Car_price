Hyperparameter tuning XGboost

```python
#Hyperparameter tuning XGboost

n_estimators = [100, 500, 1000]
learning_rate = [0.05, 0.1, 0.15, 0.2]
max_depth = [2, 3, 4]     #[2, 3, 5, 6, 7]
booster = ['gbtree']
min_child_weight = [1,2,3,4,5]
base_score = [0.25,0.5,0.75,1]
early_stopping_rounds = [10, 50, 100]

hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score,
    'early_stopping_rounds': early_stopping_rounds
    }


```

```python
import xgboost as xgb

XGreg = xgb.XGBRegressor(objective='reg:squarederror')
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import make_regression
# 4 fold
random_cv = RandomizedSearchCV(estimator=XGreg,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',
            verbose = 5,
            return_train_score = True,
            random_state=42,
            )
```

```python
random_cv.fit(X_train,y_train,eval_set=[(X_val, y_val)])
```

```python
# with open('/content/drive/MyDrive/IEEE/xgboost_model6.pkl', 'wb') as f:
#     pickle.dump(bst, f)
```

```python
random_cv.best_estimator_
```

```python
bes_reg = random_cv.best_estimator_
bes_reg
```

```python
bes_reg.fit(X_train,y_train,eval_set=[(X_val, y_val)])
```

```python
y_pred_test = bes_reg.predict(X_test)
y_pred_val = bes_reg.predict(X_val)
y_pred_train = bes_reg.predict(X_train)
```

rf model MAE = 99000


```python
mse = mean_absolute_error(y_test, y_pred_test)
print(f'Mean Absolute Error: {mse}')

mse = mean_absolute_error(y_val, y_pred_val)
print(f'Mean absolute Error: {mse}')

mse = mean_absolute_error(y_train, y_pred_train)
print(f'Mean absolute Error: {mse}')
```
